```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Delhi_v2.csv')
```

```python
data
```

```python
data=data.drop(columns=['Unnamed: 0','Landmarks','desc','Address'])
data
```

```python
data=data.fillna(0)
data
```

```python
data.hist(figsize=(15,10))
```

```python
print(data.Status.value_counts())
print(data.neworold.value_counts())
print(data.Furnished_status.value_counts())
print(data.type_of_building.value_counts())
```

```python
data=data.drop(columns=['Status','Furnished_status'])
data
```

```python
data=data.join(pd.get_dummies(data.neworold,dtype=int))
data=data.join(pd.get_dummies(data.type_of_building,dtype=int))
```

```python
data
```

```python
data=data.drop(columns=['neworold','type_of_building'])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_vars = ['price','area','latitude','longitude','Bedrooms','Bathrooms','Balcony','parking','Lift','Price_sqft']
data[num_vars] = scaler.fit_transform(data[num_vars])

X = data.drop(columns='price')
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

```python
forest.score(X_test, y_test)
```
